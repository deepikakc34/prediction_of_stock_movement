```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import time
import pprint
import pandas as pd
import numpy as np
from selenium import webdriver
```

```python
page = urlopen('https://finance.yahoo.com/quote/AAPL/news?p=AAPL').read()
soup = BeautifulSoup(page, features = 'html.parser')
posts = soup.findAll('h3', {'class':'Mb(5px)'})

news = list(posts)

urls = []
titles = []
contents = []

for new in news:
    urls.append(str(new.a['href']))
    titles.append(new.text)
    
for url in range(len(urls)):
    if 'https:/' not in urls[url]:
        urls[url] = 'https://finance.yahoo.com' + urls[url]
            
for url in range(len(urls)):
    if 'https://finance.yahoo.com/m/' in urls[url]:
        new_page = urlopen(urls[url]).read()
        new_soup = BeautifulSoup(new_page, features = 'html.parser')
        new_post = new_soup.find('div', {'class':'read-more Mt(20px) Ta(start) Ta(c)--sm'})
        new_url = new_post.a['href']
        urls[url] = new_url
        
for (url, title) in zip(urls, titles):
    if 'https://finance.yahoo.com/video/' in url:
        urls.remove(url)
        titles.remove(title)

for url in range(len(urls)):
    if 'finance.yahoo.com' in urls[url]:
        driver = webdriver.Edge(executable_path = 'C:/Users/aakul/Downloads/msedgedriver.exe')
        driver.get(urls[url])
        other_page = driver.page_source
        other_soup = BeautifulSoup(other_page, 'lxml')
        other_content = other_soup.findAll('p', {'class':'canvas-atom canvas-text Mb(1.0em) Mb(0)--sm Mt(0.8em)--sm'})
        
        content_string = ''
        
        for content in other_content:
            content_string += content.text
            
        contents.append(content_string)
        
        driver.quit()
    
    elif 'https://www.marketwatch.com' in urls[url]:
        other_post = urlopen(urls[url]).read()
        other_soup = BeautifulSoup(other_post, 'html')
        other_content = other_soup.findAll('p')
        
        content_string = ''
        
        for content in other_content:
            content_string += content.text
            
        contents.append(content_string)
    
    elif 'https://www.thestreet.com' in urls[url]:
        other_post = urlopen(urls[url]).read()
        other_soup = BeautifulSoup(other_post, 'html')
        other_content = other_soup.findAll('p')
        
        content_string = ''
        
        for content in other_content:
            content_string += content.text
            
        contents.append(content_string)
    
    elif 'https://www.barrons.com' in urls[url]:
        other_post = urlopen(urls[url]).read()
        other_soup = BeautifulSoup(other_post, 'html')
        other_content = other_soup.findAll('p')
        
        content_string = ''
        
        for content in other_content:
            content_string += content.text
            
        contents.append(content_string)
    
    elif 'https://www.investors.com' in urls[url]:
        driver = webdriver.Edge(executable_path = 'C:/Users/aakul/Downloads/msedgedriver.exe')
        driver.get(urls[url])
        other_page = driver.page_source
        other_soup = BeautifulSoup(other_page, 'lxml')
        other_content = other_soup.findAll('p')
        
        content_string = ''
        
        for content in other_content:
            content_string += content.text
            
        contents.append(content_string)
        
        driver.quit()
        
    else:
        content.append('COULD NOT SCRAPE ARTICLE CONTENT')


        
for url in urls:
    print(url)
    
print('')

for title in titles:
    print(title)
    
print('')

for content in contents:
    print(content)
```
